```python
import pandas as pd
df=pd.read_csv("/Users/Administrator/Downloads/new/IPL Matches 2008-2020.csv")
df
```

```python

df['date'] = pd.to_datetime(df['date'])


df['year'] = df['date'].dt.year


matches_per_year = df.groupby('year').size()


plt.figure(figsize=(10,6))
plt.plot(matches_per_year.index, matches_per_year.values, marker='o', linestyle='-', color='b')
plt.title('IP Matches per Year (2008-2020)')
plt.xlabel('year')
plt.ylabel('Number of Matches')
plt.grid(True)
plt.show()
```
